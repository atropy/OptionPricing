# ECDF Plots

```python
# Nifty 50
cp = pd.read_csv('/home/docboy/Desktop/Option_Pricing/datasets/CP.csv')
```

```python
def ecdf(data):
    """ Compute ECDF """
    x = np.sort(data)
    n = x.size
    y = np.arange(1, n+1) / n
    return(x,y)
```
